```python
import pandas as pd

# Read & create schedule df from excel xlsx import
schedule_df = pd.read_excel('Schedule/Complete_Schedule_2022_2023.xlsx', engine="openpyxl")

# create a list of column names
cols = ['Starters', 'MP', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', 'FT', 'FTA',
       'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS',
       '+/-', 'TS%', 'eFG%', '3PAr', 'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%',
       'STL%', 'BLK%', 'TOV%', 'USG%', 'ORtg', 'DRtg', 'BPM']

# create a new DataFrame with the columns as the index)
Visitor_df = pd.DataFrame(columns=cols)
Home_df = pd.DataFrame(columns=cols)
```

```python
# loop through the dataset
for i, row in schedule_df.iterrows():
    
    # Import Splinter and BeautifulSoup
    from splinter import Browser
    from bs4 import BeautifulSoup as soup

    # Set up Splinter
    from selenium.webdriver.chrome.service import Service
    from selenium.webdriver.chrome.options import Options
    
    # Set up Chrome options to run in headless mode
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument("--disable-extensions")
    chrome_options.add_argument('--disable-gpu')

    my_service = Service(executable_path='C://Users/bkeut/OneDrive/Desktop/chromedriver.exe')
    browser = Browser('chrome', service=my_service)
    
    # access the Game_ID, Visitor, Home, Date, and Link columns for each row
    game_id = row['Game_ID']
    visitor = row['Visitor']
    home = row['Home']
    date = row['Date']
    link = row['Link']
    
    # job running
    print(game_id, visitor, home, link)
    
    # Visit the reference site:
    url = 'https://www.basketball-reference.com/boxscores/link.html'
    url = url.replace('link', link)
    browser.visit(url)
    browser.is_element_present_by_css('div.list_text', wait_time=1)

    # Scrape the website
    html = browser.html

    # Create a BeautifulSoup object from the scraped HTML
    soup = soup(html, 'html.parser')

    # find the table containing player stats
    table1 = soup.find('table', {'id': f'box-{visitor}-game-basic'})
    table2 = soup.find('table', {'id': f'box-{visitor}-game-advanced'})
    table3 = soup.find('table', {'id': f'box-{home}-game-basic'})
    table4 = soup.find('table', {'id': f'box-{home}-game-advanced'})
    
    # convert the table to a pandas dataframe
    df1 = pd.read_html(str(table1))[0]
    df2 = pd.read_html(str(table2))[0]
    df3 = pd.read_html(str(table3))[0]
    df4 = pd.read_html(str(table4))[0]

    # clean dataframe
    df1.columns = df1.columns.droplevel(0)
    df2.columns = df2.columns.droplevel(0)
    df3.columns = df3.columns.droplevel(0)
    df4.columns = df4.columns.droplevel(0)

    # append into Team Stats
    merged_df = pd.merge(df1, df2, on=["Starters", "MP"])
    merged_df = merged_df.drop([5])
    merged_df = merged_df[merged_df["Starters"].str.contains("Team Totals") == False]
    final_df = merged_df.drop(merged_df[(merged_df.apply(lambda x: x.str.contains('Did Not Play|Did Not Dress'))).any(1)].index)
    Visitor_df = Visitor_df.append(final_df)
    
    # append into Team Stats
    merged_df = pd.merge(df3, df4, on=["Starters", "MP"])
    merged_df = merged_df.drop([5])
    merged_df = merged_df[merged_df["Starters"].str.contains("Team Totals") == False]
    final_df = merged_df.drop(merged_df[(merged_df.apply(lambda x: x.str.contains('Did Not Play|Did Not Dress'))).any(1)].index)
    Home_df = Home_df.append(final_df)
    
    #Exit Chrome Driver
    browser.quit()
    
    #Success
    print("complete")
```
